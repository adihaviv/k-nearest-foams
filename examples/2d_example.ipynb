## K-Nearest Foam 2D 

```python
import numpy as np

from field import OrthotropicField
from foam import KNearestFoam
from plot_utils import plot_2d_foam_graph

%pylab inline
```

### Define the Orthotropic Field 

```python
grid = np.array([[0, 1, 2], [0, 1, 2]])
density = np.array([[5, 5, 5], [5, 5, 5], [5, 5, 5]])
angle = np.array([[0.78, 0.78, 0.78], [0.78, 0.78, 0.78], [0.78, 0.78, 0.78]])
stretch = np.array([[(10.0, 20.0), (20.0, 40.0), (20.0, 40.0)], [(10.0, 20.0), (20.0, 40.0), (20.0, 40.0)], [(10.0, 20.0), (20.0, 40.0), (20.0, 40.0)]])

field = OrthotropicField(grid, density, angle, stretch)
```

### Create the Foam 

```python
k_foam = KNearestFoam(field, k=5)
```

```python
V, E = k_foam.vertices, k_foam.edges
```

### Visualize the Foam 

```python
plot_2d_foam_graph(k_foam)
```
