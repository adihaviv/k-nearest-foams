### Define the Orthotropic Field 

```python
grid = np.array([
    [0, 1, 2], 
    [0, 1, 2, 3, 4, 5, 6, 7]
])

d = 15
density = np.array([
    [d ,d ,d ,d ,d ,d ,d, d],
    [d ,d ,d ,d ,d ,d ,d, d],
    [d ,d ,d ,d ,d ,d ,d, d]
])

alpha = 2 * 0.78
beta = 0.78
angle = np.array([
    [beta, beta, alpha, alpha, alpha, alpha, beta, beta], 
    [beta, beta, alpha, alpha, alpha, alpha, beta, beta],
    [beta, beta, alpha, alpha, alpha, alpha, beta, beta]
])

s1 = (10.0, 30.0)
s2 = (10.0, 30.0)
stretch = np.array([
    [s1, s1, s1, s2, s2, s2, s1, s1], 
    [s1, s1, s1, s2, s2, s2, s1, s1],
    [s1, s1, s1, s2, s2, s2, s1, s1]
])

field = OrthotropicField(grid, density, angle, stretch)
```

### Create the Foam 

```python
k_foam = KNearestFoam(field, k=2)
```

```python
V, E = k_foam.vertices, k_foam.edges
```

### Visualize the Foam 

```python
plot_2d_foam_graph(k_foam)
```
